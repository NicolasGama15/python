# 2.  **Explicación Conceptual Detallada**


**¿Qué es la Visualización de Datos en Python?**

Es el proceso de crear representaciones gráficas de la información utilizando librerías de Python. Estas representaciones pueden ser desde gráficos de líneas y barras simples (estáticos) hasta complejos mapas de calor o diagramas 3D interactivos.

**Las Tres Librerías Clave:**

1.  **`matplotlib`**: Es la "madre" de todas las librerías de visualización en Python. Es extremadamente potente y personalizable hasta el más mínimo detalle. A veces, su sintaxis puede ser un poco extensa, pero te da control total. Piensa en ella como una caja de herramientas completa para construir cualquier gráfico desde cero.
2.  **`seaborn`**: Está construida sobre `matplotlib`. Su objetivo es hacer que los gráficos estadísticos sean más fáciles de crear y visualmente más atractivos por defecto. Con menos código, puedes generar gráficos complejos y estéticamente agradables. Es como un set de plantillas de diseño para `matplotlib`.
3.  **`plotly`**: Es la estrella de la interactividad. Mientras que `matplotlib` y `seaborn` producen principalmente imágenes estáticas, `plotly` genera gráficos interactivos que puedes explorar: hacer zoom, pasar el ratón para ver valores, filtrar datos, etc. Son ideales para dashboards y páginas web.

**Conceptos Clave:**
*   **Figura (`Figure`)**: Es el lienzo o contenedor principal donde se dibuja todo. Un solo `Figure` puede contener múltiples gráficos.
*   **Ejes (`Axes`)**: Es el gráfico individual en sí (el área con los ejes X e Y, las marcas, las etiquetas). Una `Figure` puede tener uno o más `Axes`.
*   **Marcas (`Marks`)**: Son los elementos visuales que representan los datos (puntos, barras, líneas).
*   **Estética (`Aesthetics`)**: Son los atributos visuales que codifican la información (color, tamaño, forma).

**Errores Comunes:**
*   **Olvidar `plt.show()`**: En scripts de Python (no siempre en Jupyter, que a veces lo muestra automáticamente), si no usas `plt.show()`, el gráfico se crea en memoria pero nunca se muestra en pantalla.
*   **Datos en formato incorrecto**: La mayoría de las librerías funcionan mejor con estructuras de datos de `pandas` (como DataFrames) o `numpy` (arrays). Pasar una lista simple a veces puede dar resultados inesperados.

#### Ejemplo 1: Comparando Ventas Mensuales (Matplotlib vs. Seaborn)

Este ejemplo muestra cómo crear el mismo gráfico de barras con ambas librerías para que aprecies sus diferencias.


````python
# Celda 2: Creación de datos de ventas
data = {
    'Mes': ['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun'],
    'Ventas': [150, 180, 220, 190, 250, 270]
}
df_ventas = pd.DataFrame(data)

# --- Gráfico con Matplotlib ---
plt.figure(figsize=(10, 4)) # Definimos el tamaño de la figura

plt.subplot(1, 2, 1) # Creamos un subgráfico (1 fila, 2 columnas, posición 1)
plt.bar(df_ventas['Mes'], df_ventas['Ventas'], color='skyblue')
plt.title('Ventas Mensuales (Matplotlib)')
plt.ylabel('Ventas (en miles)')
plt.xlabel('Mes')
plt.xticks(rotation=45) # Rotamos las etiquetas del eje X para que no se superpongan

# --- Gráfico con Seaborn ---
plt.subplot(1, 2, 2) # Creamos el segundo subgráfico (posición 2)
sns.barplot(x='Mes', y='Ventas', data=df_ventas, palette='viridis')
plt.title('Ventas Mensuales (Seaborn)')
plt.ylabel('Ventas (en miles)')
plt.xlabel('Mes')
plt.xticks(rotation=45)

plt.tight_layout() # Ajusta los gráficos para que no se solapen
plt.show()
````

#### Ejemplo 2: Explorando la Distribución de Datos (Histograma y Boxplot con Seaborn)

Vamos a generar datos aleatorios que simulen las notas de un examen y visualizaremos su distribución.

````python
# Celda 3: Generación de datos de notas y visualización
np.random.seed(42) # Para que los resultados aleatorios sean reproducibles
notas_examen = np.random.normal(loc=75, scale=10, size=200).astype(int) # 200 notas con media 75 y desv. estándar 10

# --- Visualización de la distribución ---
plt.figure(figsize=(12, 5))

# Histograma: Muestra la frecuencia de las notas
plt.subplot(1, 2, 1)
sns.histplot(notas_examen, kde=True, color='salmon') # kde=True añade una línea de densidad
plt.title('Distribución de Notas de Examen (Histograma)')
plt.xlabel('Nota')
plt.ylabel('Frecuencia')

# Boxplot: Muestra la mediana, cuartiles y valores atípicos
plt.subplot(1, 2, 2)
sns.boxplot(x=notas_examen, color='lightblue')
plt.title('Resumen Estadístico de Notas (Boxplot)')
plt.xlabel('Nota')

plt.show()
````

#### Ejemplo 4: Matriz de Correlación con un Mapa de Calor (Seaborn)

Una tarea fundamental en el análisis de datos es entender cómo las variables numéricas se relacionan entre sí. Un mapa de calor (heatmap) de una matriz de correlación es la herramienta perfecta para esto.

Usaremos el dataset `penguins`, que contiene medidas de diferentes especies de pingüinos.

````python
# Celda 5: Mapa de calor de correlaciones

# 1. Cargar el dataset 'penguins' de Seaborn
df_penguins = sns.load_dataset('penguins')

# 2. Calcular la matriz de correlación
# .corr() solo funciona con columnas numéricas, pandas las seleccionará automáticamente.
# Redondeamos a 2 decimales para que sea más legible.
corr_matrix = df_penguins.corr(numeric_only=True).round(2)

# 3. Crear la figura y los ejes con Matplotlib para tener más control
plt.figure(figsize=(8, 6))

# 4. Crear el mapa de calor con Seaborn
sns.heatmap(
    corr_matrix, 
    annot=True,      # annot=True muestra los valores numéricos en cada celda
    cmap='coolwarm', # cmap define la paleta de colores. 'coolwarm' es ideal para correlaciones (azul=negativo, rojo=positivo)
    vmin=-1,         # vmin y vmax aseguran que la barra de color vaya de -1 a 1
    vmax=1
)

# 5. Añadir un título y mostrar el gráfico
plt.title('Mapa de Calor de Correlación - Dataset Penguins', fontsize=15)
plt.show()
````

#### Ejemplo 5: Comparando Distribuciones con Gráficos de Violín (Seaborn)

Un gráfico de violín es como un boxplot, pero con esteroides. No solo muestra el resumen estadístico (mediana, cuartiles), sino que también muestra la distribución de la densidad de los datos. Es perfecto para comparar cómo se distribuye una variable numérica a través de diferentes categorías.

Usaremos el dataset `titanic` para comparar la edad de los pasajeros según la clase en la que viajaban.

````python
# Celda 6: Gráficos de Violín para comparar distribuciones

# 1. Cargar el dataset 'titanic'
df_titanic = sns.load_dataset('titanic')

# 2. Crear la figura
plt.figure(figsize=(12, 7))

# 3. Crear el gráfico de violín con Seaborn
sns.violinplot(
    data=df_titanic,
    x='class',          # Variable categórica en el eje X
    y='age',            # Variable numérica en el eje Y
    hue='sex',          # Dividimos cada violín por otra categoría (sexo)
    split=True,         # 'split=True' une los violines de 'hue' en uno solo para una comparación directa
    palette='pastel',
    order=['First', 'Second', 'Third'] # Especificamos el orden de las categorías
)

# 4. Añadir títulos y etiquetas para mayor claridad
plt.title('Distribución de Edad por Clase y Sexo en el Titanic', fontsize=16)
plt.xlabel('Clase del Pasajero', fontsize=12)
plt.ylabel('Edad', fontsize=12)
plt.show()
````

#### Ejemplo 6: Animación Interactiva con Gráficos de Burbuja (Plotly)

Este es un ejemplo avanzado que muestra el verdadero poder de Plotly: la animación. Recrearemos el famoso gráfico de Hans Rosling que muestra la evolución de la esperanza de vida vs. el PIB per cápita a lo largo del tiempo.

Usaremos el dataset `gapminder`, que está convenientemente incluido en Plotly.

````python
# Celda 7: Gráfico de burbuja animado e interactivo

# 1. Cargar el dataset de ejemplo 'gapminder' de Plotly
df_gapminder = px.data.gapminder()

# 2. Crear la figura animada con Plotly Express
fig = px.scatter(
    df_gapminder,
    x="gdpPercap",                  # Eje X: PIB per cápita
    y="lifeExp",                    # Eje Y: Esperanza de vida
    animation_frame="year",         # Variable que controla la animación (el tiempo)
    animation_group="country",      # Variable para identificar los objetos entre frames (para que las burbujas no se 'teletransporten')
    size="pop",                     # El tamaño de la burbuja representa la población
    color="continent",              # El color de la burbuja representa el continente
    hover_name="country",           # Lo que se muestra en negrita al pasar el ratón
    log_x=True,                     # Usamos una escala logarítmica para el eje X, ya que el PIB varía enormemente
    size_max=60,                    # Tamaño máximo de la burbuja más grande para que no sea excesivo
    range_y=[25,90]                 # Rango fijo para el eje Y para que la comparación entre años sea justa
)

# 3. Personalizar el título
fig.update_layout(title_text='Evolución de Esperanza de Vida vs. PIB per Cápita (1952-2007)')

# 4. Mostrar la figura interactiva
fig.show()
````

### **Matplotlib (`plt` y `ax`) - El Control Preciso**
Es la base de todo. Ideal cuando necesitas personalizar cada mínimo detalle del gráfico. La mejor práctica es usar la **interfaz orientada a objetos (`ax`)**, que es más clara y potente.

#### **Dos Interfaces (¡Importante!)**

```python
# 1. Interfaz Pyplot (similar a MATLAB, buena para plots rápidos)
import matplotlib.pyplot as plt
plt.plot([1,2,3], [4,5,1])
plt.show()

# 2. Interfaz Orientada a Objetos (OO) - ¡LA RECOMENDADA!
# Te da control sobre la figura (lienzo) y los ejes (gráficos) por separado.
fig, ax = plt.subplots() # Crea una figura (fig) y un conjunto de ejes (ax)
ax.plot([1,2,3], [4,5,1])
plt.show()
```

sns.histplot(data=df, x='edad', kde=True, bins=20) # Histograma con curva de densidad

sns.barplot(data=df, x='region', y='ventas') # Muestra una agregación (media por defecto)

# --- Gráficos Matriciales (visualizar matrices de datos) ---

### **Plotly Express (`px`) - Interactividad Instantánea**
Crea gráficos interactivos con una sintaxis muy sencilla y legible. Ideal para exploración de datos y dashboards.

#### **Sintaxis Fundamental**
La estructura es casi siempre la misma: `px.tipo_de_grafico(data_frame, ...argumentos...)`

#### **Argumentos Mágicos (Semántica similar a Seaborn)**
*   `color`: Columna para diferenciar por color.
*   `symbol`: Columna para diferenciar por forma del marcador.
*   `size`: Columna para diferenciar por tamaño del marcador.
*   `hover_name`: Columna cuyo valor aparece en negrita al pasar el ratón.
*   `hover_data`: Lista de columnas para mostrar información extra al pasar el ratón.
*   `facet_row` / `facet_col`: Columna para crear una cuadrícula de subplots.
*   `animation_frame`: Columna para crear animaciones (e.g., año).
*   `title`: Título del gráfico.

#### **Tipos de Gráficos Comunes**
```python
import plotly.express as px

fig = px.scatter(df, x='total_bill', y='tip', color='smoker', facet_col='day')
fig = px.bar(df, x='day', y='total_bill', color='sex')
fig = px.line(df_stocks, x='date', y='price', color='ticker')
fig = px.histogram(df, x='total_bill', nbins=50, marginal='rug') # marginal='rug' o 'box'
fig = px.box(df, x='day', y='tip', color='smoker')
fig = px.sunburst(df, path=['continente', 'pais'], values='poblacion') # Gráfico jerárquico
fig = px.choropleth(df, locations='iso_alpha', color='gdp', hover_name='country') # Mapas
```

#### **Personalización Post-Creación (`fig.update_...`)**
Se usa para refinar el gráfico después de crearlo con `px`.

```python
# Actualizar el layout general (títulos, ejes, fuentes, leyenda)
fig.update_layout(
    title_text='Título Principal del Gráfico',
    xaxis_title='Nuevo Título Eje X',
    yaxis_title='Nuevo Título Eje Y',
    legend_title='Título de la Leyenda',
    font=dict(family="Arial", size=12)
)

# Actualizar las trazas (los datos visuales: líneas, marcadores)
# Por ejemplo, cambiar el estilo de los marcadores en un scatter plot
fig.update_traces(
    marker=dict(size=12, line=dict(width=2, color='DarkSlateGrey')),
    selector=dict(mode='markers') # Aplica solo a trazas que tienen marcadores
)
```
